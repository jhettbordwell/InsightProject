```python
import pandas as pd
import numpy as np
import glob as glob
```

```python
def map2meds(condMeds, df):
    medDF = df.drop(columns=[col for col in df if col not in condMeds])
    medications = ['']*len(df)
    for med in condMeds:
        inds = medDF[medDF[med].eq(1)].index
        for ind in inds: 
            medications[ind] = med
        
    return medications
    
    
def Label500(conditions):
    listOdf = []
    for condition in conditions:
        data_file = 'Final_processed_reviews/{:s}_processed.csv'.format(condition)
        df = pd.read_csv(data_file, sep='$', index_col=0)
        np.random.seed = 616
        inds = np.arange(0,len(df))
        np.random.shuffle(inds)
        inds = inds[:100]
        df = df.loc[inds]
        df = df.reset_index()
        
        meds_file = 'UniqueMedications/Medications_unique_{:s}.csv'.format(condition)
        medications = [med.strip() for med in pd.read_csv(meds_file, sep='$')['Medication']]
        
        medCol = map2meds(medications, df)
        reviews = df['Full Review']
        condCol = [condition]*len(medCol)
        listOdf.append(pd.DataFrame(np.array([condCol, medCol, reviews]).T, 
                       columns=['Condition', 'Medication', 'Full Review']))
        
    fulldf = pd.concat(listOdf, axis=0).reset_index().drop(columns='index')
    return fulldf
    
conditions = ['ADHD', 'Anxiety', 'Bipolar-Disorder', 'Depression', 'Schizophrenia']
combo = Label500(conditions)
```
